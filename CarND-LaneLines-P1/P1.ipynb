## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images directory.
def build_lane_finding_pipeline(image):
    #Grayscale the image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 3
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold) 

    # This time we are defining a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),(490,310), (490, 310), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 15     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 40 #minimum number of pixels making up a line
    max_line_gap = 20    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    lines_image = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)    
    
    # Add the lines to the initial image
    lines_edges = weighted_img(lines_image, image, =0.8, =1., =0.) 
    return lines_edges

for image_name in os.listdir("test_images/"):
    #Read and process test images
    img = mpimg.imread("test_images/" + image_name)
    img_processed = build_lane_finding_pipeline(img)
    
    #Create directory and output images' name
    img_directory = "test_images_output/"
    img_name_output = image_name[:-4] + "_output.jpg"
    
    #Output the processed images 
    if not os.path.exists("test_images_output/"):
        os.makedirs(img_directory)
        mpimg.imsave(img_directory + img_name_output, img_processed)
    else:
        mpimg.imsave(img_directory + img_name_output, img_processed)
        
    #Show these images
    plt.imshow(img_processed)
    plt.show()
````

````python
#Create a list to stroe previous frames
previous_frame_list = []

# Change draw_lines function and ues it in pipeline
def draw_lines_connect(img, lines, color=[255, 0, 0], max_y=720, min_y=330, thickness=15):
    """
    I will modify the 'draw_lines' function of Helper Functions in order to connect the separated line segments together in the form of a line.
    """
 
    #Firstly, extract left x ,y and right x,y separately and stroe in the lists
    #Create stroe lists which represent the segments of left and right lines
    left_x1 = []
    left_y1 = []
    left_x2 = []
    left_y2 = []
    
    right_x1 = []
    right_y1 = []
    right_x2 = []
    right_y2 = []   
    
    #Use the y_max and y_min to store the endpoit of y coordinates of two lines
    #max_y = 720 #img.shape[0] 
    #min_y = 330 #img.shape[0] 
    
    #Add the x,y values to suitable lists
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 -y1) / (x2 - x1) #Find the slope of the solely line
            #Positive slope of the right line
            if slope > 0.5 and slope < 0.8:
                right_x1.append(x1)
                right_x2.append(x2)
                right_y1.append(y1)
                right_y2.append(y2)
            #Nagative slope of the left line
            elif slope < -0.5 and slope > -0.8:
                left_x1.append(x1)
                left_x2.append(x2)
                left_y1.append(y1)
                left_y2.append(y2)
            #Get the min value of y
            #min_y = min(min_y, y1, y2)
                
    
    #Secondly, calculate the average num of x,y coordinates of left and right lines
    #Get the left averge x,y
    avg_left_x1 = (sum(left_x1) / float(len(left_x1))) if float(len(left_x1)) !=0 else 0   
    avg_left_x2 = (sum(left_x2) / float(len(left_x2))) if float(len(left_x2)) !=0 else 0
    avg_left_y1 = (sum(left_y1) / float(len(left_y1))) if float(len(left_y1)) !=0 else 0
    avg_left_y2 = (sum(left_y2) / float(len(left_y2))) if float(len(left_y2)) !=0 else 0
    
    #Get the right averge x,y
    avg_right_x1 = (sum(right_x1) / float(len(right_x1))) if float(len(right_x1)) !=0 else 0
    avg_right_x2 = (sum(right_x2) / float(len(right_x2))) if float(len(right_x2)) !=0 else 0
    avg_right_y1 = (sum(right_y1) / float(len(right_y1))) if float(len(right_y1)) !=0 else 0
    avg_right_y2 = (sum(right_y2) / float(len(right_y2))) if float(len(right_y2)) !=0 else 0
    
    
    #Thirdly, calculate the slope and intercept for left and right direction through those avg numbers
    #Initialize the slope and Get the slope
    avg_left_slope = 0
    avg_right_slope = 0
    avg_left_intercept = 0
    avg_right_intercept = 0
    
    if ((avg_left_x2 - avg_left_x1)!=0) & ((avg_right_x2 - avg_right_x1) !=0):
        avg_left_slope = (avg_left_y2 - avg_left_y1) / (avg_left_x2 - avg_left_x1)
        avg_right_slope = (avg_right_y2 - avg_right_y1) / (avg_right_x2 - avg_right_x1)
    
        #Get the intercept
        avg_left_intercept = avg_left_y2 - avg_left_slope * avg_left_x2
        avg_right_intercept = avg_right_y2 - avg_right_slope * avg_right_x2
        
        #Add the slope and intercption into the previous_frame_list
        if len(previous_frame_list) <= 5:
            previous_frame_list.append([avg_left_slope, avg_right_slope, avg_left_intercept, avg_right_intercept])
        else:
            previous_frame_list.pop()
            previous_frame_list.append([avg_left_slope, avg_right_slope, avg_left_intercept, avg_right_intercept])
    
    # Compute te averge number of slope and intercption       
    if len(previous_frame_list) >0:
        previous_frame_array = np.array(previous_frame_list)
        avg_previous = previous_frame_array.mean(axis = 0)
        avg_left_slope = avg_previous[0]
        avg_right_slope = avg_previous[1]
        avg_left_intercept = avg_previous[2]
        avg_right_intercept = avg_previous[3]
        
        
    #Fouthly, find the endpoint to define the whole straight line which has the average slope and intercept
    #Initialize the points and Find the end points
    left_min_y1 = 0
    left_min_x1 = 0
    left_max_y2 = 0
    left_max_x2 = 0

    right_min_y1 = 0
    right_min_x1 = 0
    right_max_y2 = 0
    right_max_x2 = 0
    
    #Find the left end points
    if (avg_left_slope != 0) & (avg_right_slope != 0):
        left_min_y1 = min_y
        left_min_x1 = int((left_min_y1 - avg_left_intercept) / avg_left_slope)
        left_max_y2 = max_y
        left_max_x2 = int((left_max_y2 - avg_left_intercept) / avg_left_slope)
    
        #Find the right end points
        right_min_y1 = min_y
        right_min_x1 = int((right_min_y1 - avg_right_intercept) / avg_right_slope)
        right_max_y2 = max_y
        right_max_x2 = int((right_max_y2 - avg_right_intercept) / avg_right_slope)
    
    
    #Fifthly, draw the right and left line using the end points
    cv2.line(img, (left_min_x1, left_min_y1), (left_max_x2, left_max_y2), color, thickness)
    cv2.line(img, (right_min_x1, right_min_y1), (right_max_x2, right_max_y2), color, thickness)
 

    
    
         
def hough_lines_connect(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """    
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    #Call the changed function draw_lines_connect
    draw_lines_connect(line_img, lines)
    return line_img
 
    

def build_lane_finding_pipeline_extened(image):
    #Grayscale the image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 3
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)

    # This time we are defining a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(0,imshape[0]),(490,310), (490, 310), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 50     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 100 #minimum number of pixels making up a line
    max_line_gap = 100    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    lines_image = hough_lines_connect(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)    
    
    # Add the lines to the initial image
    lines_edges = weighted_img(lines_image, image, =0.8, =1., =0.) 
    return lines_edges


#Read the image respectively and process it
for image_name in os.listdir("test_images/"):
    #Read and process test images
    img = mpimg.imread("test_images/" + image_name)
    img_processed = build_lane_finding_pipeline_extened(img)
    
    #Create directory and output images' name
    img_directory = "test_images_output_extended/"
    img_name_output = image_name[:-4] + "_output.jpg"
    
    #Output the processed images 
    if not os.path.exists(img_directory):
        os.makedirs(img_directory)
        mpimg.imsave(img_directory + img_name_output, img_processed)
    else:
        mpimg.imsave(img_directory + img_name_output, img_processed)
        
    #Show these images
    plt.imshow(img_processed)
    plt.show()
````

````python
#Create a list to stroe previous frames
previous_frame_list = []

# Change draw_lines_connect function and ues it in pipeline for the challenge problem
def draw_lines_refined(img, lines, color=[255, 0, 0], max_y=720, min_y=330, thickness=15):
    """
    I will modify the 'draw_lines' function of Helper Functions in order to connect the separated line segments together in the form of a line.
    """
 
    #Firstly, extract left x ,y and right x,y separately and stroe in the lists
    #Create stroe lists which represent the segments of left and right lines
    left_x1 = []
    left_y1 = []
    left_x2 = []
    left_y2 = []
    
    right_x1 = []
    right_y1 = []
    right_x2 = []
    right_y2 = []   
    
    #Use the y_max and y_min to store the endpoit of y coordinates of two lines
    #max_y = 720 #img.shape[0] 
    #min_y = 330 #img.shape[0] 
    
    #Add the x,y values to suitable lists
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 -y1) / (x2 - x1) #Find the slope of the solely line
            #Positive slope of the right line
            if slope > 0.5 and slope < 0.8:
                right_x1.append(x1)
                right_x2.append(x2)
                right_y1.append(y1)
                right_y2.append(y2)
            #Nagative slope of the left line
            elif slope < -0.5 and slope > -0.8:
                left_x1.append(x1)
                left_x2.append(x2)
                left_y1.append(y1)
                left_y2.append(y2)
            #Get the min value of y
            #min_y = min(min_y, y1, y2)
                
    
    #Secondly, calculate the average num of x,y coordinates of left and right lines
    #Get the left averge x,y
    avg_left_x1 = (sum(left_x1) / float(len(left_x1))) if float(len(left_x1)) !=0 else 0   
    avg_left_x2 = (sum(left_x2) / float(len(left_x2))) if float(len(left_x2)) !=0 else 0
    avg_left_y1 = (sum(left_y1) / float(len(left_y1))) if float(len(left_y1)) !=0 else 0
    avg_left_y2 = (sum(left_y2) / float(len(left_y2))) if float(len(left_y2)) !=0 else 0
    
    #Get the right averge x,y
    avg_right_x1 = (sum(right_x1) / float(len(right_x1))) if float(len(right_x1)) !=0 else 0
    avg_right_x2 = (sum(right_x2) / float(len(right_x2))) if float(len(right_x2)) !=0 else 0
    avg_right_y1 = (sum(right_y1) / float(len(right_y1))) if float(len(right_y1)) !=0 else 0
    avg_right_y2 = (sum(right_y2) / float(len(right_y2))) if float(len(right_y2)) !=0 else 0
    
    
    #Thirdly, calculate the slope and intercept for left and right direction through those avg numbers
    #Initialize the slope and Get the slope
    avg_left_slope = 0
    avg_right_slope = 0
    avg_left_intercept = 0
    avg_right_intercept = 0
    
    if ((avg_left_x2 - avg_left_x1)!=0) & ((avg_right_x2 - avg_right_x1) !=0):
        avg_left_slope = (avg_left_y2 - avg_left_y1) / (avg_left_x2 - avg_left_x1)
        avg_right_slope = (avg_right_y2 - avg_right_y1) / (avg_right_x2 - avg_right_x1)
    
        #Get the intercept
        avg_left_intercept = avg_left_y2 - avg_left_slope * avg_left_x2
        avg_right_intercept = avg_right_y2 - avg_right_slope * avg_right_x2
        
        #Add the slope and intercption into the previous_frame_list
        if len(previous_frame_list) <= 6:
            previous_frame_list.append([avg_left_slope, avg_right_slope, avg_left_intercept, avg_right_intercept])
        else:
            previous_frame_list.pop()
            previous_frame_list.append([avg_left_slope, avg_right_slope, avg_left_intercept, avg_right_intercept])
    
    # Compute te averge number of slope and intercption       
    if len(previous_frame_list) >0:
        previous_frame_array = np.array(previous_frame_list)
        avg_previous = previous_frame_array.mean(axis = 0)
        avg_left_slope = avg_previous[0]
        avg_right_slope = avg_previous[1]
        avg_left_intercept = avg_previous[2]
        avg_right_intercept = avg_previous[3]
        
        
    #Fouthly, find the endpoint to define the whole straight line which has the average slope and intercept
    #Initialize the points and Find the end points
    left_min_y1 = 0
    left_min_x1 = 0
    left_max_y2 = 0
    left_max_x2 = 0

    right_min_y1 = 0
    right_min_x1 = 0
    right_max_y2 = 0
    right_max_x2 = 0
    
    #Find the left end points
    if (avg_left_slope != 0) & (avg_right_slope != 0):
        left_min_y1 = min_y
        left_min_x1 = int((left_min_y1 - avg_left_intercept) / avg_left_slope)
        left_max_y2 = max_y
        left_max_x2 = int((left_max_y2 - avg_left_intercept) / avg_left_slope)
    
        #Find the right end points
        right_min_y1 = min_y
        right_min_x1 = int((right_min_y1 - avg_right_intercept) / avg_right_slope)
        right_max_y2 = max_y
        right_max_x2 = int((right_max_y2 - avg_right_intercept) / avg_right_slope)
    
    
    #Fifthly, draw the right and left line using the end points
    cv2.line(img, (left_min_x1+45, left_min_y1), (left_max_x2, left_max_y2), color, thickness)
    cv2.line(img, (right_min_x1, right_min_y1), (right_max_x2, right_max_y2), color, thickness)
    
    
    
def hough_lines_refined(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """    
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    #Call the changed function draw_lines_connect
    draw_lines_refined(line_img, lines, max_y=720, min_y=450, thickness=15)
    return line_img

#Chage the parameters to fit the curved lines
def build_lane_finding_pipeline_refined(image):
    #Grayscale the image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 1
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define our parameters for Canny and apply
    low_threshold = 50
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)

    # This time we are defining a four sided polygon to mask
    imshape = image.shape
    #print(image.shape)
    vertices = np.array([[(0,imshape[0]),(490,310), (490, 310), (imshape[1],imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 50     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 100 #minimum number of pixels making up a line
    max_line_gap = 100    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    lines_image = hough_lines_refined(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)    
    
    # Add the lines to the initial image
    lines_edges = weighted_img(lines_image, image, =0.8, =1., =0.) 
    return lines_edges
````
